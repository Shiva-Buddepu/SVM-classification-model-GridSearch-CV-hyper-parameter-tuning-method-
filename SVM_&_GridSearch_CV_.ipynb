```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/breast_cancer.csv')
df.head()
```

```python
df.outcome.value_counts()
```

```python
df.shape
```

```python
#x and y split
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.outcome
y.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#model building
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred1=model.predict(x_train)
```

```python
#evaluate the model
from sklearn.metrics import accuracy_score,classification_report
print("test accuracy: ",accuracy_score(y_test,y_pred))
print("train accuracy: ",accuracy_score(y_train,y_pred1))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
#recall (our aim is to get nearer to 1)
37/(37+7)
```

```python
print(classification_report(y_test,y_pred))
```

#hyper parameter tuning

```python
#using grid search CV
from sklearn.model_selection import GridSearchCV
parameters={'kernel':['linear','rbf'],
            'C':[0.1,0.2,0.5,1.0],
            'gamma':['scale','auto']}
```

```python
clf= GridSearchCV(model,param_grid=parameters,verbose=2) #verbose print step by step
clf.fit(x_train,y_train)
```

```python
clf.best_score_ #max accuracy of testing the data
```

```python
clf.best_params_ #these parameters give max accuracy
```

```python
#now using best parameters in model
model1=SVC(C=1,gamma='scale',kernel='linear')
model1.fit(x_train,y_train)
```

```python
y_ =model1.predict(x_test)
```

```python
y_1 =model1.predict(x_train)
```

```python
print("testing accuracy",accuracy_score(y_test,y_))
print("traing accuracy",accuracy_score(y_train,y_1))
```

```python
pd.crosstab(y_test,y_)
```

```python
#now recall of 1 :
43/(43+1)
```

```python
print(classification_report(y_test,y_))
```
